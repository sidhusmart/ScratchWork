```
import pandas as pd
import numpy as np
data_result = pd.read_csv('/content/bert_output/test_results.tsv', sep='\t')
data = pd.read_csv('/content/data/test_compare.tsv', sep='\t')
final_data = pd.concat([data, data_result], axis=1)
final_data.columns = ['id','label','reviewText','probNeg','probPos']
final_data['pred'] = np.where(final_data['probPos'] > 0.5, 1, 0)

from sklearn.metrics import accuracy_score
accuracy_score(final_data['label'],final_data['pred'])
```
